```python
from selenium import webdriver
import time
from bs4 import BeautifulSoup
```

```python
browser = webdriver.Firefox()   
    
for i in range(0,300,5):
    #Creates the URLs needed to access the data; the i variable is used 
    #to continue to the next page based on the link set-up
    url2 = 'https://www.tripadvisor.com/Hotel_Review-g60978-d102730-Reviews-or' + str(i) + '-Mill_Street_Inn-Newport_Rhode_Island.html'
    #Gets the url2 variable, which is the webpage we need
    browser.get(url2)
        #Lets the page load for 5 seconds.
    time.sleep(5)
        #Identifies all of the clickable links in order to obtain the full reviews.
    element_list = browser.find_elements_by_xpath("//span[@class='taLnk ulBlueLinks']")
        #Iteration clicks all of the 'More' links. The 'try' statement allows the iteration 
        #to continue with 'pass' when an error message appears-caused by TA.
    for e in element_list:
        try:
            e.click()
        except:
            pass
                    #Variable to get the page source through BeautifulSoup.
        html = browser.page_source
                    #Creates a soup object. 
        soup = BeautifulSoup(html, "lxml")
                    #Finds the review sections needed.
        reviews = soup.find_all('div', 'reviewSelector')
                    #Iterates through the review sections, pulls the rating & review, and writes them into the CSV file.
    for r in reviews:
        print("Rating: ", int(r.find('span','ui_bubble_rating')['class'][1].split('_')[1])/10)
        print("Review: ", r.p.text)
            
```

```python
# Recode the Ratinds from 1-3 stars to negative and 4-5 stars to positive
z=[]
divbox = soup.find_all("div","ui_column is-9")
print(divbox)
    
for i in divbox:
    if "ui_bubble_rating bubble_10" in str(i):
        z.append('neg')
    elif "ui_bubble_rating bubble_20" in str(i):
        z.append('neg')
    elif "ui_bubble_rating bubble_30" in str(i):
        z.append('neg')
    elif "ui_bubble_rating bubble_40" in str(i):
        z.append('pos')
    elif "ui_bubble_rating bubble_50" in str(i):
        z.append('pos')
    else:
        print("uk") 
```

```python
from nltk.tokenize import word_tokenize
import string
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
```

```python
#cleaned the data
Reviews = []
for r in soup.find_all('div', 'reviewSelector'):
    
    Review = r.p.text
    words = word_tokenize(Review)
    clean_words = [word.lower() for word in words if word not in set(string.punctuation)]
    english_stops = set(stopwords.words('english'))
    clean_words = [word for word in clean_words if word not in english_stops]
    wordnet_lemmatizer = WordNetLemmatizer()
    lemma_list = [wordnet_lemmatizer.lemmatize(word) for word in clean_words]
    reviews.append(lemma_list)
    print(lemma_list)
```

```python
combined = zip(reviews,z)
```

```python
print(len(reviews))
```

```python
def bag_of_words(words):
    return dict([(word, True) for word in words])
```

```python
train_data = []
for r, v in combined:
    bag_of_words(r)
    train_data.append((bag_of_words(r),v))
    
print(train_data)
```

```python
import random
random.shuffle(train_data)
print(len(train_data))
```

```python
train_set, train_set = train_data[0:2], train_data[2:]
```

```python
import nltk
import collections
from nltk.metrics.scores import (accuracy, precision, recall, f_measure)
from nltk.classify import NaiveBayesClassifier
nb_classifier = NaiveBayesClassifier.train(train_set)
nb_classifier.most_informative_features(n=5)
```
